```python
x = [1,4,5,8,2,3]
y = [23,46,30,50,2,6]
plt.plot(x,y)
plt.show()
```

```python
# Method of Plotting
```

```python
# Method 1
```

```python
main = plt.figure()
sub_parts = main.add_subplot()
sub_parts.plot(x,y)
plt.show()
```

```python
# Method 2
```

```python
main = plt.figure()
sub_parts = main.add_axes([1,1,1,1])
sub_parts.plot(x,y)
plt.show()
```

```python
# Method 3
```

```python
main,sub_parts = plt.subplots()
sub_parts.plot(x,y)
plt.show()
```

```python
 # Labeling on Plot
```

```python
main,sub_parts = plt.subplots()
sub_parts.plot(x,y)
sub_parts.set(title = "Testing Plot", xlabel = "Time", ylabel= "Speed")
plt.show()
```
