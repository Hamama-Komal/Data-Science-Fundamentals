```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [2,10,6,7,8]
y = [11,13,15,17,18]
```

```python
# 4 Types of Plot: 1- Line Plot,  2- Scatter Plot, 3- Bar Plot & 4- Histogram
```

```python
# Line Plot
```

```python
fig, subplot = plt.subplots()
subplot.plot(x,y);
```

```python
fig, subplot = plt.subplots()  
subplot.plot(x,np.sin(x));      # Apply sin function
```

```python
fig, subplot = plt.subplots()
subplot.plot(x,np.exp(x));   # applying exponetial function
```

```python
# Scatter Plot
```

```python
fig, subplot = plt.subplots()
subplot.scatter(x,y);
```

```python
# Bar Plot
```

```python
data = {'item1': 30,'item2': 10,'item3': 70,'item4': 90}
fig, subplot = plt.subplots()
subplot.bar(data.keys(), data.values());   # show vertical bars
```

```python
data = {'item1': 30,'item2': 10,'item3': 70,'item4': 90}
fig, subplot = plt.subplots()
subplot.barh(list(data.keys()), list(data.values()));       # Shows horizontal bars
```

```python
# Histogram
```

```python
fig, subplot = plt.subplots()
subplot.hist(x);
```

```python
# Gussain Distribution or Normal Distribution: Graph of random data is always bell curve
```

```python
ran_data = np.random.randn(10000)
fig, subplot = plt.subplots()
subplot.hist(ran_data);
```

```python
# Multiple Plot
```

```python
fig, ((part1, part2), (part3, part4)) = plt.subplots(nrows = 2, ncols = 2, figsize = [5,5])
part1.plot(x,y)
part2.bar(data.keys(), data.values());
part3.scatter(x,y);
part4.hist(x);
```
