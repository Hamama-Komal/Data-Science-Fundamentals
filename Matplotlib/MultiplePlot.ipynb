```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# 1- prepare the data
x = [1,4,5,8,2,3]
y = [23,46,30,50,2,6]
```

```python
# 2- prepare the plot
main,sub_parts = plt.subplots()
# 3- set plot enviroment
sub_parts.plot(x,y)
# 4- customization
sub_parts.set(title = "Testing Plot", xlabel = "Time", ylabel= "Speed")
# 5- showing the garph
plt.show()
```
