```python
# For Data Preparation
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
# For Modeling
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
np.random.seed(7)

# Prepare Data
data = pd.read_csv("Missing Heart Attack Data.csv")
# data.dtypes
# data.isna().sum()

# Data Transformation:

float_features = ['chol', 'Idl', 'hdl', 'bp', 'smoking', 'diabetes']
float_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='mean'))])

preprocessing = ColumnTransformer(transformers=[('float', float_transformer, float_features)])

# Creating model pipeline
model = Pipeline(steps=[('preprocessing', preprocessing), ('model', RandomForestRegressor())])

# Split Data
x = data.drop('heart_attack', axis=1)
y = data['heart_attack']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# Fit Model

model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
# Setting Hyper-parameters
hyper_params = {
    'preprocessing__float__imputer__strategy': ['mean','median'],
    'model__n_estimators': [90,100, 110],
    'model__min_samples_split': [5,6,7],
    'model__min_samples_leaf': [3,4,5],
    'model__max_features': ['sqrt', 'auto'],
    'model__max_depth': [None, 1, 10]
}

gscv_model = GridSearchCV(model, hyper_params, cv=3, verbose = 2)
gscv_model.fit(x_train, y_train)
```

```python
gscv_model.score(x_test, y_test)
```
