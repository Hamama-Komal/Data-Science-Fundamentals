```python
# Hyper-parameters
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.get_params()     # shows the hyper parameters of the models
```

```python
# 3 methods to change or update hyper-parameter:
# 1- Manually
# 2- Randomly with RandomSearchCV
# 3- Exhautively GridSearch

# To Tune the model, we spilt the data into 3 parts now test, train and validation.
```

```python
import pandas as pd
import numpy as np

heart_data = pd.read_csv("Heart Attack Data.csv")
mixed_heart_data = heart_data.sample(frac=1)
mixed_heart_data.head()
```

```python
# spilt data into X & Y
x = mixed_heart_data.drop('heart_attack', axis = 1)
y = mixed_heart_data['heart_attack']
# spilting into 3 parts: 70%, 15% & 15%
train_split = round(0.7 * len(mixed_heart_data))
vaild_split = round(train_split + 0.15 * len(mixed_heart_data))
x_train, y_train = x[:train_split], y[:train_split]
x_valid, y_valid = x[train_split:vaild_split], y[train_split:vaild_split]
x_test, y_test = x[vaild_split:], y[vaild_split:]

len(mixed_heart_data), len(x_train), len(x_valid), len(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evalute_hyper_parameters(y_true, y_predicted):
    accuracy = accuracy_score(y_true, y_predicted)
    precision = precision_score(y_true, y_predicted)
    recall = recall_score(y_true, y_predicted)
    f1 = f1_score(y_true, y_predicted)

    print(f'Accuracy: {accuracy}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1: {f1}')
    
    metric_dic = {'Accuracy': round(accuracy, 2), 'Precision': round(precision, 2), 'Recall': round(recall, 2), 'F1': round(f1, 2)}

    return metric_dic
```

```python
clf.fit(x_train, y_train)
y_predicted = clf.predict(x_valid)
simple_results = evalute_hyper_parameters(y_valid, y_predicted)
```

```python
# Tunning -Hyperparameters:

# 1- n_estimators

np.random.seed(7)
clf2 = RandomForestClassifier(n_estimators=1000)
clf2.fit(x_train, y_train)
y_predicted = clf2.predict(x_valid)
estimator_results = evalute_hyper_parameters(y_valid, y_predicted)
```

```python

# 1- max_depth

np.random.seed(7)
clf3 = RandomForestClassifier(max_depth = 100)
clf3.fit(x_train, y_train)
y_predicted = clf2.predict(x_valid)
max_depth_results = evalute_hyper_parameters(y_valid, y_predicted)
```

```python
# Tunning Hyper-parameters using Randomized Search CV (RSCV)
hy_para = {
    'max_depth': [None, 5, 10, 30, 50, 100],
    'n_estimators': [10, 100, 200, 500, 1000],
    'min_samples_split': [2,4,6],
    'min_samples_leaf': [1,2,4],
    'max_features': ["auto", "sqrt"]
}

x = mixed_heart_data.drop('heart_attack', axis = 1)
y = mixed_heart_data['heart_attack']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

clf = RandomForestClassifier()

from sklearn.model_selection import RandomizedSearchCV

rscv_clf = RandomizedSearchCV(estimator = clf, param_distributions= hy_para, n_iter = 10, cv = 5, verbose = 2)

rscv_clf.fit(x_train, y_train)
```

```python
rscv_clf.best_params_
```

```python
rscv_predicted_y = rscv_clf.predict(x_test)
rscv_predicted_y
```

```python
rscv_results = evalute_hyper_parameters(y_test, rscv_predicted_y)
```

```python
# Flaw of RSCV is it test half combination.
```

```python
# Tunning Hyper-parameters using Grid Search Cross validation (GSCV):

best_hyper_params = {
    'n_estimators': [90,100, 110],
    'min_samples_split': [5,6,7],
    'min_samples_leaf': [3,4,5],
    'max_features': ['sqrt', 'auto'],
    'max_depth': [None, 1, 10]
}

from sklearn.model_selection import GridSearchCV

gscv_clf = GridSearchCV(estimator = clf, param_grid= best_hyper_params,  cv = 5, verbose = 2)

gscv_clf.fit(x_train, y_train);
```

```python
gscv_clf.best_params_
```

```python
gscv_predicted_y = gscv_clf.predict(x_test)
gscv_results = evalute_hyper_parameters(y_test, gscv_predicted_y)
```

```python
# Result Comparision:

all_results = {
    'Normal': simple_results,
    'Estimator': estimator_results,
    'Max Depth': max_depth_results,
    'RSCV': rscv_results,
    'GSCV': gscv_results
}

comparsion = pd.DataFrame(all_results)
comparsion
```
