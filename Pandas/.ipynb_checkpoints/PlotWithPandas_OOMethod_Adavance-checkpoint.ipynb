```python
import pandas as pd
```

```python
data = pd.read_csv('test_dataset.csv')
data
```

```python
type(data['ram'][0])
```

```python
data['total ram'] = data['ram'].cumsum()       # cumsum() : Cumulative Sum
data
```

```python
data['dates'] = pd.date_range('01/01/2015', periods = len(data))     # To add date columns
data
```

```python
data.plot(x = 'dates', y = 'total ram', kind = 'scatter');
```

```python
data.plot(x = 'ram', y = 'total ram', kind = 'scatter');
```

```python
data.plot(x = 'ram', y = 'dates', kind = 'scatter');
```
