```python
import matplotlib.pyplot as plt
```

```python
fig, room = plt.subplots(figsize=(6,5))
data.plot(x = 'ram', y = 'battery_power', kind = 'scatter', c='dual_sim', ax = room);    #3 OO method
```

```python
# Lifecycle of OO Method
```

```python
fig, ax = plt.subplots(figsize=(6,5))  # prepare the area
scatter_plot = ax.scatter(x = data['ram'], y = data['battery_power'], c = data['dual_sim']);  # plot the data
ax.set(xlabel = 'Ram', ylabel = 'Battery Power', title = 'Dual Sim Suspectibilty') # customization
ax.legend(*scatter_plot.legend_elements(), title = 'Dual Sim')  # legend
scatter_plot
```

```python
# Mutliple Plot
```

```python
fig, (plot1, plot2) = plt.subplots(nrows = 2, ncols = 1, figsize=(6,5), sharex = True)  # prepare the area for 2 plot  # sharex >> show share the axis

# plot 1
scatter_plot = plot1.scatter(x = data['ram'], y = data['battery_power'], c = data['dual_sim'], cmap = 'cool');  # cmap >> color map, adding different colors
plot1.set( ylabel = 'Battery Power', title = 'Dual Sim Suspectibilty')
plot1.legend(*scatter_plot.legend_elements(), title = 'Dual Sim') 
plot1.axhline(y = data['battery_power'].mean(), linestyle = "--")  # to show the mean value

# plot 2
scatter_plot = plot2.scatter(x = data['ram'], y = data['clock_speed'], c = data['dual_sim'], cmap = 'plasma'); 
plot2.set(xlabel = 'Ram', ylabel = 'Clock Speed') 
plot2.legend(*scatter_plot.legend_elements(), title = 'Dual Sim') 
plot2.axhline(y = data['clock_speed'].mean(), linestyle = "--")  # to show the mean value

plt.style.use('classic')  # use to add the theme on plot >> to use the 1 core theme, define it on the top of file
plt.suptitle('Mobile Phone Information', fontsize = 20, fontweight= 'bold')      # use to add the main heading with styling option
scatter_plot;

fig.savefig('Mobile Phone Information.png') # use to save the plot
```
